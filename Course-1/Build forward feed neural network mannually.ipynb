```python
import numpy as np
import matplotlib.pyplot as plt
```

Step 1: Generate data 

```python
NClass = 500 # 500 Samples per class
X1 = np.random.randn(NClass,2)+np.array([0,-2])
X2 = np.random.randn(NClass,2)+np.array([2,2])
X3 = np.random.randn(NClass,2)+np.array([-2,2])
Y = np.array([0]*NClass+[1]*NClass+[2]*NClass)
X= np.vstack([X1,X2,X3])

## Create a data set with two features and one label
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,s=100,alpha=0.5)
plt.show()
```

Step 2: Build Neural Network Mannually

```python
D = 2 # Number of Hidden Layers
M = 3 # Number of Neuron in each layer
K = 3 # Number of Class
# Initial Weights by randomizing them
W1 = np.random.randn(D,M)
b1 = np.random.randn(M)
W2 = np.random.randn(M,K)
b2 = np.random.randn(K)
```

Build Neural Network Mannually (Theory is understandable)

```python
def forward(X,W1,b1,W2,b2):
    Z = 1/(1+np.exp(-X.dot(W1)-b1))
    A = Z.dot(W2)+b2
    expA = np.exp(A)
    Y=expA/expA.sum(axis=1,keepdims=True)
    return Y
```

```python
from sklearn.metrics import accuracy_score
```

Make prediction

```python
Prob = forward(X,W1,b1,W2,b2)
```

```python
Pred = np.argmax(Prob,axis=1)
```

```python
accuracy_score(Pred,Y)
```
